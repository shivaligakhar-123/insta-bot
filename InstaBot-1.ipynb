```python
# seeting up all the imports
# please take care of responsiveness while checking this project
# if the page will break then classes will change and that will break the code
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(ChromeDriverManager().install())
import time
from selenium.common.exceptions import NoSuchElementException
```

## Login to your Instagram Handle
### Submit with sample username and password

```python
# logging into account
wait = WebDriverWait(driver, 10)
driver.get("https://www.instagram.com/")
fields = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_2hvTZ")))
username,password = fields[0],fields[1]
username.clear()
# add username here
username.send_keys("SAMPLE_USERNAME")
password.clear()
# add password here
password.send_keys("SAMPLE_PASSWORD")
login = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"sqdOP")))[1]
login.click()
time.sleep(3) #used when connection is slow
# login successful
notnow = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'cmbtv')]/button")))
notnow.click()
time.sleep(3) #used when connection is slow
notinotnow = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"HoLwm")))
notinotnow.click()
```

## Like/Unlike posts
### Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
### Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
# searching for dilsefoodie
search = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"XTCLo")))
search.clear()
search.send_keys("dilsefoodie")
dilsefoodie_page=wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"yCE8d")))[0]
dilsefoodie_page.click()
```

```python
# Likes
# clicking on first post
post1 = driver.find_elements_by_class_name("_9AhH0")[0]
post1.click()
# then going till 30 pages using next button
for i in range(30):
    like = wait.until(EC.presence_of_element_located((By.XPATH,"//section[contains(@class,'ltpMr')]/span/button")))
    svg = driver.find_element_by_css_selector(".eo2As > section >span >button > div >span>svg")
#     this condition is checking if post is already liked or not
    if svg.get_attribute("aria-label")=="Unlike":
        print(i+1,"post is already liked")
    else:
        print(i+1,"post is liked")
        like.click()
    next_button=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_65Bje.coreSpriteRightPaginationArrow")))
    next_button.click()
# then closing the posts by clicking close button    
close = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'yiMZG')]/button")))
close.click()
```

## Extract list of followers
### Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
# creating a function for extracting 500 followers of a page, value takes the name of page
def follower_list(value):   
    driver.get("https://www.instagram.com/"+value+"/")
    follow=wait.until(EC.presence_of_element_located((By.XPATH,"//ul[contains(@class,'k9GMp')]/li[2]/a")))
    follow.click()
    time.sleep(3)
    usernames=driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")
#     making loop run till number of followers be 500
    while len(usernames)<500:
        scroll = driver.execute_script('return document.querySelector(".isgrP").scrollHeight;')
        scroll_to = driver.execute_script('document.querySelector(".isgrP").scrollTo(0,arguments[0]);',scroll)
        # time.sleep(3)       Used when net connection is slow
        usernames=driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")
# finding 500 usernames
    usernames=driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")[:500]
    i=0
# printing them
    for ele in usernames:
        print(i+1,end="\t")
        link = ele.get_attribute("href")
        print(link.split("/")[-2])
        i+=1
    close=wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'WaOAr')]/button")))
    close.click()
```

## Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
### If You have already seen the story.
### Or The user has no story.
### Or View the story if not yet seen.

```python
driver.get("https://www.instagram.com/coding.ninjas/")
```

```python
# For Windows User

story_check =  wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'XjzKX')]/div")))
# applying condition for user has no story
if story_check.get_attribute("class") == 'RR-M- ':
    print("No story")
# if user has a story
else:
    story=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"CfWVH")))
# if height == 210 then story is not seen
    if story.get_attribute("height")=="210":
        print("Story is not seen")
        print("Watching stories Now")
        i=0
#         applying loop for watching story
        while story.get_attribute("height")=="210":
            story_check =  wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'XjzKX')]/div")))
            story_check.click()
            close_cross = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "coreSpriteCloseLight")))
            close_cross.click()
            print( str(i+1) + "part of story seen")
            i=i+1
            story=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"CfWVH")))
        print("All Stories Seen")


# if height ==208 then story is seen
    elif story.get_attribute("height")=="208":
        print("Story is already seen")
```
